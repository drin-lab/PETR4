```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("PETR4.csv", parse_dates = True, index_col= "Date")
```

```python
df.head()
```

```python
df4 = df['Close']
```

```python
from matplotlib import axes
```

```python
fig, ax = plt.subplots(figsize=(13, 6))
plt.style.use("ggplot")
df4.plot(ax=ax)

# Add labels to the plot
style = dict(size=10, color='gray')

ax.text('2016-4', 14.5, "R$ 13,57 - IMPEACHMENT", **style,
       bbox=dict(facecolor='white', alpha=0.8, edgecolor='black', boxstyle='round'))
ax.text('2016-4', 16, "AGOSTO 2016", **style,
       bbox=dict(facecolor='white', alpha=0.8, edgecolor='black', boxstyle='round'))


ax.text('2017-5', 13, "R$ 12,37 - JOESLEY DAY", **style,
         bbox=dict(facecolor='white', alpha=0.8, edgecolor='black', boxstyle='round'))
ax.text('2017-5', 14.5, "MAIO 2017", **style,
       bbox=dict(facecolor='white', alpha=0.8, edgecolor='black', boxstyle='round'))

ax.text('2018-5', 18 , "R$ 17,19 - GREVE DOS CAMINHONEIROS", **style, 
       bbox=dict(facecolor='white', alpha=0.8, edgecolor='black', boxstyle='round'))
ax.text('2018-5', 19.5, "MAIO 2018", **style,
       bbox=dict(facecolor='white', alpha=0.8, edgecolor='black', boxstyle='round'))

ax.text('2019-7', 11 , "R$ 12,07 - CORONAVIRUS", **style, 
       bbox=dict(facecolor='white', alpha=0.8, edgecolor='black', boxstyle='round'))
ax.text('2019-7', 12.5, "MARÇO 2020", **style,
       bbox=dict(facecolor='white', alpha=0.8, edgecolor='black', boxstyle='round'))


# Label the axes
ax.set(title='Variação de preço PETR4 (período 2016/2020)', ylabel='Variação de preço (R$)')
plt.xlabel("")

plt.gcf()
plt.savefig("petr_2020-5.jpg", dpi = 100)
plt.show()
```
